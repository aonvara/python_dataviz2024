##### **ข้อที่ 1**. ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

###### วิเคราะห์โจทย์: จากข้อมูลข้างต้นจะเห็นว่าที่ในหนึ่งเดือนจะทำการเก็บตัวอย่างวัดคุณภาพน้ำอยู่ 3 rep. ในแต่ละ station ทั้งหมด 6 stations ดังนั้นการหาค่าเฉลี่ยของ Chol_a จะประกอบด้วย 6 stations

###### **วิธีทำ**

```python
#ตัดเฉพาะ column ที่ต้องการ
smaller_df_1 = water_quality_df[['month','station', 'Chlo_a']]

#ตัดข้อมูลที่มีค่า NaN จากตาราง
smaller_df_1 = smaller_df_1.dropna()
smaller_df_1.shape

smaller_df_1.head()
```

```python
#สร้าง column ที่มีเฉพาะปีเพื่อใช้ในการแบ่งข้อมูล

#ใช้คำสั่ง split() แบ่ง '_' ที่ข้อมูลใน column 'month' นำเฉพาะตัวเลขมาสร้างใหม่เป็น column 'year'
smaller_df_1['year'] = smaller_df_1['month'].str.split('_').str[1].astype(int)

smaller_df_1
```

```python
#แยกคิดเฉพาะข้อมูลที่อยู่ในปี 2018
smaller_df_1_2018 = smaller_df_1[smaller_df_1['year'] == 2018]
```

```python
#หาค่าเฉลี่ยของ Chlo_a แต่ละ station จากตาราง smaller_df_1_2018 โดยใช้ group by
mean_chlo_a_by_station_2018 = smaller_df_1_2018.groupby('station')['Chlo_a'].mean()
mean_chlo_a_by_station_2018
```

```python
#แยกคิดเฉพาะข้อมูลที่อยู่ในปี 2019
smaller_df_1_2019 = smaller_df_1[smaller_df_1['year'] == 2019]
```

```python
#หาค่าเฉลี่ยของ Chlo_a แต่ละ station จากตาราง smaller_df_1_2019 โดยใช้ group by column 'station'
mean_chlo_a_by_station_2019 = smaller_df_1_2019.groupby('station')['Chlo_a'].mean()
mean_chlo_a_by_station_2019
```

###### **คำตอบ**

```python
#หาส่วนต่างระหว่างค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019
mean_chlo_a_diff = mean_chlo_a_by_station_2018 - mean_chlo_a_by_station_2019
print(f' ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ระหว่างปี 2018 และ ปี 2019 ทั้ง 6 stations ประกอบด้วย')
mean_chlo_a_diff
```

##### **ข้อที่ 2**. สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

```python
#ตัดเฉพาะ column ที่ต้องการนำมาคำนวณ
smaller_df_2 = water_quality_df[['month','station', 'Chlo_b']]

#ตัดข้อมูลที่มีค่า NaN จากตาราง
smaller_df_2 = smaller_df_2.dropna()
smaller_df_2.shape

smaller_df_2.head()
```

###### วิเคราะห์โจทย์: หาค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือนแยกปีกัน


```python
#ใช้คำสั่ง split() แบ่ง '_' ที่ข้อมูลใน column 'month'
#นำเฉพาะตัวอักษรมาสร้างใหม่เป็น column 'month' แทนที่ column เดิม และนำตัวเลขมาสร้างใหม่เป็น column 'year'
smaller_df_2['year'] = smaller_df_2['month'].str.split('_').str[1].astype(int)
smaller_df_2
```

###### ในปี 2018

```python
#เลือกเฉพาะข้อมูลที่เป็นปี 2018 มา
smaller_df_2_2018 = smaller_df_2[smaller_df_2['year'] == 2018]

smaller_df_2_2018
```

```python
#หาค่าเฉลี่ยของ Chlo_b แต่ละ station จากตาราง smaller_df_1_2018 โดยใช้ group by() column 'month' และ 'station'
mean_Chlo_b_by_month_station_2018 = smaller_df_2_2018.groupby(['month','station'])['Chlo_b'].mean()
mean_Chlo_b_by_month_station_2018
```

###### ในปี 2019

```python
#เลือกเฉพาะข้อมูลที่เป็นปี 2019 มา
smaller_df_2_2019 = smaller_df_2[smaller_df_2['year'] == 2019]

smaller_df_2_2019
```

```python
#หาค่าเฉลี่ยของ Chlo_b แต่ละ station จากตาราง smaller_df_1_2019 โดยใช้ group by() column 'month' และ 'station'
mean_Chlo_b_by_month_station_2019 = smaller_df_2_2019.groupby(['month','station'])['Chlo_b'].mean()
mean_Chlo_b_by_month_station_2019
```

###### สร้างตาราง

```python
#นำตารางค่าเฉลี่ย Chol_a ที่หาได้มาเปลี่ยน column เป็น rowname โดยใช้ unstack()
#2018
answer_df_mean_Chlo_b_2018 = mean_Chlo_b_by_month_station_2018.unstack()

#2019
answer_df_mean_Chlo_b_2019 = mean_Chlo_b_by_month_station_2019.unstack()
```

###### คำตอบ

```python
print(f' ตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือนในปี 2018')
answer_df_mean_Chlo_b_2018
```

```python
print(f' ตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือนในปี 2019')
answer_df_mean_Chlo_b_2019
```
