###### **เริ่มสืบค้นข้อมูลจากทั้งพยานคนที่ 1 และ 2**


---

1. หมายเลขสมาชิกบนกระเป๋าขึ้นต้นด้วย "48Z"
2. สมาชิกระดับโกลด์
3. ออกกำลังกายเมื่อสัปดาห์ที่แล้วในวันที่ 9 มกราคม
4. ป้ายทะเบียนรถที่มี "H42W"

```python
#นำเข้าข้อมูลที่เกี่ยวข้อง
#ข้อมูลสมาชิก ฟิตเนส Get Fit Now
path_to_gym_member = '/content/drive/MyDrive/data_viz_2024_DATA/Exam_2/get_fit_now_member.csv'
gym_member_df = pd.read_csv(path_to_gym_member)
gym_member_df.head()
```

```python
#นำเข้าข้อมูลที่เกี่ยวข้อง
#ข้อมูลเช็คอิน ฟิตเนส Get Fit Now
path_to_gym_checkin = '/content/drive/MyDrive/data_viz_2024_DATA/Exam_2/get_fit_now_check_in.csv'
gym_checkin_df = pd.read_csv(path_to_gym_checkin)
gym_checkin_df.head()
```

```python
#นำเข้าข้อมูลที่เกี่ยวข้อง
#ข้อมูลใบขับขี่
path_to_drivers = '/content/drive/MyDrive/data_viz_2024_DATA/Exam_2/drivers_license.csv'
drivers_df = pd.read_csv(path_to_drivers)
drivers_df.head()
```

```python
#หาว่าใครมีสมาชิกฟิตเนสเป็นระดับ gold และมีรหัสสมาชิกขึ้นต้นด้วย 48Z
Suspect_murder_gym_1 = gym_member_df[(gym_member_df['membership_status'] == 'gold') & (gym_member_df['id'].str.startswith('48Z'))]
Suspect_murder_gym_1
```

```python
#หาว่าใครเข้าใช้ฟิตเนตวันที่ 9 มกราคม 2018 และมีรายชื่อตรงกับสมาชิกฟิตเนสเป็นระดับ gold และมีรหัสสมาชิกขึ้นต้นด้วย 48Z
#พบว่าทั้งสองคนเข้าฟิตเนสในวันที่ 9 มกราคม เหมือนกัน ดังนั้นยังไม่สามารถสรุปตัวฆาตรกรได้
Suspect_murder_gym_2 = gym_checkin_df[(gym_checkin_df['check_in_date'] == 20180109) & (gym_checkin_df['membership_id'].isin(Suspect_murder_gym_1['id']))]
Suspect_murder_gym_2
```

```python
#หาว่ารถคันไหนที่มีป้ายทะเบียนรถประกอบด้วย "H42W"
Suspect_murder_drivers = drivers_df[drivers_df['plate_number'].str.contains('H42W')]
Suspect_murder_drivers
```

```python
#หารายละเอียดของ Joe Germuska และ Jeremy Bowers ในข้อมูลประชากร
#เนื่องจากข้อมูล id ในข้อมูลใบขับขี่ จะตรงกับข้อมูล license id ในข้อมูลประชากร

Suspect_murder_person = person_df[(person_df['name'] == 'Joe Germuska') | (person_df['name'] == 'Jeremy Bowers')]
Suspect_murder_person
```

```python
#พบว่า Jeremy Bowers ที่มี license_id = 423327 ตรงกับ id ในข้อมูลใบขับขี่ ที่มีป้ายทะเบียนรถที่ประกอบด้วย H42W
Suspect_murder_person_drivers = Suspect_murder_person[Suspect_murder_person['license_id'].isin(Suspect_murder_drivers['id'])]
Suspect_murder_person_drivers
```

###### **สรุปรายชื่อผู้ต้องสงสัยจากทั้งพยานคนที่ 1 และ 2**


---

1. หมายเลขสมาชิกบนกระเป๋าขึ้นต้นด้วย "48Z" = Joe Germuska และ Jeremy Bowers
2. สมาชิกระดับโกลด์ = Joe Germuska และ Jeremy Bowers
3. ออกกำลังกายเมื่อสัปดาห์ที่แล้วในวันที่ 9 มกราคม = Joe Germuska และ Jeremy Bowers
4. ป้ายทะเบียนรถที่มี "H42W" = Jeremy Bowers
