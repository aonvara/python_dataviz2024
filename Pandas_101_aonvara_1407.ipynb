## **1. นิทานที่ยาวที่สุด และสั้นที่สุด (ไม่นับสัญลักษณ์พิเศษ)**

````
#เนื้อเรื่องของนิทานจะอยู่ในหัวข้อ description
#และจะนับแค่ตัวอักษรเท่านั้น ไม่นับสัญลักษณ์พิเศษ
folk_df['description']
````

````
#ก่อนนำสัญลักษณ์ออก
#max
folk_df['description'].str.len().max()
````

````
#min
folk_df['description'].str.len().min()
````

````
#พิมพ์ prompt ใน gemini ว่า "วิธีเลือกแค่อักษรภาษาไทยในตาราง"
#ถ้า error กดที่อธิบายข้อผิดพลาด

import re

def keep_thai_only(text):
  if isinstance(text, str): # Check if the input is a string
    return ''.join(re.findall(r'[\u0E00-\u0E7F]', text))
  else:
    return '' # Return an empty string if not a string

folk_df['thai_only'] = folk_df['description'].apply(keep_thai_only)
````

````
folk_df['thai_only']
````

````
#หลังนำสัญลักษณ์ออก
#max
max = folk_df['thai_only'].str.len().max()
````

````
#min
min = folk_df['thai_only'].str.len().min()
````

##**ตอบ**

````
print(f'นิทานที่ยาวที่สุดคือ {max} ตัวอักษร')
print(f'นิทานที่สั้นที่สุดคือ {min} ตัวอักษร')
````

````
#พิมพ์ prompt ใน gemini ว่า "วิธีหาว่าโค้ดนี้ max = folk_df['thai_only'].str.len().max() อยู่ตำแหน่งใดในตาราง"
# หาตำแหน่งของ max
max_length = folk_df['thai_only'].str.len().max()
max_index = folk_df['thai_only'].str.len().idxmax()

# แสดงผลลัพธ์
print("ตำแหน่งของ min:", max_index)
print("ค่าของ min:", max_length)

# แสดงข้อมูลในแถวที่มีค่า max
folk_df.loc[[max_index]]
````

````
# หาตำแหน่งของ min
min_length = folk_df['thai_only'].str.len().min()
min_index = folk_df['thai_only'].str.len().idxmin()

# แสดงผลลัพธ์
print("ตำแหน่งของ min:", min_index)
print("ค่าของ min:", min_length)

# แสดงข้อมูลในแถวที่มีค่า min
folk_df.loc[[min_index]]
````

### การวนลูป record ในตาราง  (```.iterrows```)
