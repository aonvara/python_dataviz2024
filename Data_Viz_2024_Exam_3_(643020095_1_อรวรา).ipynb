#### วิเคราะห์โจทย์: comments ใดที่มีจำนวนไลค์และการตอบกลับเป็นจำนวนมาก แสดงให้เห็นว่า comments ดังกล่าวมีผู้คนที่มีความรู้สึกและประสบการณ์ร่วมด้วยเป็นจำนวนมาก เท่ากับว่าประสบการณ์โรคมะเร็งที่เจ้าของ comments กล่าวอาจมีประเด็นบางอย่างที่น่าสนใจ

```python
#หา comment ที่มียอดไลค์จำนวนมาก
#จะเห็นว่าเมื่อเรียงลำดับจะมีค่าโดดขึ้นมาอยู่ 3 ค่า
Top_Like = Data_Docter_df.sort_values(by='ยอดไลค์', ascending=False).head(5)
Top_Like
```

```python
#หา comment ที่มียอดการตอบกลับจำนวนมาก
#จะเห็นว่าเมื่อเรียงลำดับจะมีค่าโดดขึ้นมาอยู่ 1 ค่า
Top_Response = Data_Docter_df.sort_values(by='จำนวนการตอบกลับ', ascending=False).head(5)
Top_Response
```

```python
#จะเห็นว่าเป็น comment ของ Nubeer Kithiya ที่มีค่าโดดจาก comment อื่นๆทั้งยอดไลค์และการตอบกลับ
#ซึ่งรายละเอียด comment ได้กล่าวไว้ดังนี้
Data_Docter_df[Data_Docter_df['name'] == 'Nubeer Kithiya']['comments'].values[0]
```

### **จุดที่ 1 ที่น่าสนใจในข้อมูลนี้**
การที่ comments ของ ***Nubeer Kithiya*** ที่มีค่าโดดจาก comments อื่นๆทั้งยอดไลค์และการตอบกลับ มีประเด็นที่น่าสนใจคือการเป็นมะเร็งลำไส้จากพฤติกรรมในชีวิตประจำวันที่ดูไม่ปกติ คือ "ถ้าไม่ใช่บ้าน หรือที่คุ้นเคย จะไม่ยอมถ่ายเลย" ทำให้อาจเกิดประเด็นถกเถียงภายใน comments เกี่ยวกับพฤติกรรมที่อาจมีผู้รู้สึกไม่เห็นด้วย/แปลก หรือรู้สึกแบบเดียวกัน

#### วิเคราะห์โจทย์: จากจุดที่ 1 อาจเป็นไปได้ว่าการที่ยอดไลค์เยอะ จะทำให้การตอบกลับเยอะตาม

```python
#คัดเอาเฉพาะ column ยอดไลค์ และ การตอบกลับ
Like_Response = Data_Docter_df[['ยอดไลค์', 'จำนวนการตอบกลับ']]

#เปลี่ยนชื่อ column
Like_Response = Like_Response.rename(columns={'ยอดไลค์': 'Like', 'จำนวนการตอบกลับ': 'Response'})
Like_Response
```

```python
pd.plotting.scatter_matrix(Like_Response);
```
