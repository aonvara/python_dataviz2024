```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Use pd.read_pickle to read a pickle file
final_data_tuple = pd.read_pickle('/content/drive/MyDrive/data_viz_2024_DATA/Xnn_ppm.pickle')

# Assuming the tuple structure is (data, metadata), extract the data part
final_data = final_data_tuple[0]  # Assuming the data is the first element of the tuple


# Check the type of final_data
print(type(final_data))

# If final_data is a NumPy array, convert it to a DataFrame
if isinstance(final_data, np.ndarray):
    # You might need to adjust the column names if needed
    final_data = pd.DataFrame(final_data)
    #If you have column names in a separate list, you can pass it to the DataFrame constructor:
    #final_data = pd.DataFrame(final_data, columns=your_column_names)

print(final_data.shape)
final_data.head()

# Check if the DataFrame is empty
if final_data.empty:
    print("Error: Input DataFrame is empty. Please check your data loading process.")
else:
    # Assuming you want to reduce to 2 principal components for visualization
    n_components = 2

    # Create a PCA object
    pca = PCA(n_components=n_components)

    # Fit and transform the data
    principal_components = pca.fit_transform(final_data)

    # Create a new DataFrame with the principal components
    principal_df = pd.DataFrame(data = principal_components, columns = ['principal component 1', 'principal component 2'])

    # Visualize the data in the reduced dimension space
    plt.figure(figsize=(8, 6))
    plt.scatter(principal_df['principal component 1'], principal_df['principal component 2'])
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('PCA of your data')
    plt.show()

    # You can also check the explained variance ratio of the principal components
    print(pca.explained_variance_ratio_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans # Import KMeans for clustering

# ... (Your existing code for loading and preprocessing 'final_data') ...

# Check if the DataFrame is empty
if final_data.empty:
    print("Error: Input DataFrame is empty. Please check your data loading process.")
else:
    # Assuming you want to reduce to 2 principal components for visualization
    n_components = 2

    # Create a PCA object
    pca = PCA(n_components=n_components)

    # Fit and transform the data
    principal_components = pca.fit_transform(final_data)

    # Create a new DataFrame with the principal components
    principal_df = pd.DataFrame(data=principal_components, columns=['principal component 1', 'principal component 2'])

    # Apply KMeans clustering to the principal components
    kmeans = KMeans(n_clusters=3)  # You can adjust the number of clusters
    principal_df['cluster'] = kmeans.fit_predict(principal_df[['principal component 1', 'principal component 2']])

    # Visualize the data in the reduced dimension space with cluster labels
    plt.figure(figsize=(8, 6))
    plt.scatter(principal_df['principal component 1'], principal_df['principal component 2'], c=principal_df['cluster'], cmap='viridis')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('PCA of your data with KMeans Clustering')
    plt.colorbar(label='Cluster')
    plt.show()

    # You can also check the explained variance ratio of the principal components
    print(pca.explained_variance_ratio_)

    # Create a table showing the cluster assignments for each data point
    cluster_table = pd.concat([final_data, principal_df['cluster']], axis=1)
    cluster_table
```
