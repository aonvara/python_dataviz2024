# การแสดงผลการกระจายของข้อมูล

````
import pandas as pd
from matplotlib import pyplot as plt
````

````
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df.head()
````

## Scatter 2D

## scatter - default

````
plt.scatter(df['PetalLength'],df['PetalWidth'])
````

### color

````
plt.scatter(df['PetalLength'],df['PetalWidth'],c='g')
````

## ใส่สีให้แต่ละจุด

````
set(df['Name']) #ดูว่ามีค่าอะไรบ้าง
````

````
df2 = df.replace({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica':2})
df2
````

````
plt.scatter(df['PetalLength'],df['PetalWidth'],c=df2['Name'])
````

### alpha กำหนดความโปร่งแสง

````
plt.scatter(df['PetalLength'],df['PetalWidth'],c=df2['Name'],alpha=0.5)
````

## ใช้ขนาดของ marker แสดงค่าของ feature

````
plt.scatter(df['SepalLength'],df['PetalWidth'],s=50*df['PetalWidth'],c=df2['Name'],alpha=0.5)
````

## ตัวแปร Marker ใช้ในการเปลี่ยนหน้าตาของ marker

````
plt.scatter(df['PetalLength'],df['PetalWidth'],s=100,c=df2['Name'],alpha=0.5,marker='*')
````

## การ plot กราฟหลายๆอันซ้อนกัน

````
plt.scatter(df['PetalLength'][:50],df['PetalWidth'][:50],s=100,c='m',alpha=0.5,marker='$S$')
plt.scatter(df['PetalLength'][50:100],df['PetalWidth'][50:100],s=100,c='c',alpha=0.5,marker='$E$')
plt.scatter(df['PetalLength'][100:],df['PetalWidth'][100:],s=100,c='r',alpha=0.5,marker='$I$')
````

###Label-Legend

```plt.plot()``` input  ตัวที่ 3 คือ ตัวกำหนดหน้าตาของ marker

'หน้าตาmarker'-'ลักษณะของเส้นเชื่อมจุด'-'สี'

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'or',alpha=0.5)
````

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'o-r',alpha=0.5)
````

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'o:r',alpha=0.5)
````

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'or',alpha=0.5,label='setosa')
plt.plot(df['PetalLength'][50:100],df['PetalWidth'][50:100],'xc',alpha=0.5,label='versicolor')
plt.plot(df['PetalLength'][100:],df['PetalWidth'][100:],'*m',alpha=0.5,label='virginica')
plt.legend(loc='lower right')
````

### การใส่ชื่อกราฟ และชื่อแกน

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'or',alpha=0.5,label='setosa')
plt.plot(df['PetalLength'][50:100],df['PetalWidth'][50:100],'xc',alpha=0.5,label='versicolor')
plt.plot(df['PetalLength'][100:],df['PetalWidth'][100:],'*m',alpha=0.5,label='virginica')
plt.legend(loc='lower right')
plt.title('Iris flower data distribution') ##ชื่อกราฟ
plt.xlabel('PetalLength') ##ชื่อแกน X
plt.ylabel('PetalWidth'); ##ชื่อแกน y
````

### 3D scatter

````
ax = plt.axes(projection ="3d")
ax.scatter3D(df['PetalLength'],df['PetalWidth'],df['SepalWidth'],c=df2['Name'])
````

````
genres_list = [
    'Action', 'Adventure', 'Animation', 'Biography', 'Comedy',
    'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy',
    'History', 'Horror', 'Musical', 'Mystery', 'Romance',
    'Sci-Fi', 'Sport', 'Thriller', 'War', 'Western'
]
````

````
def split_genres(row):
  # Check if the genre value is a string before splitting
  if isinstance(row['Genre'], str):
    genres = row['Genre'].split(',')
    return [genre.strip() for genre in genres]
  else:
    # Handle cases where the genre is not a string (e.g., return an empty list or NaN)
    return []

movie_df['Genre'] = movie_df.apply(split_genres, axis=1)
````

````
import pandas as pd
from matplotlib import pyplot as plt
from google.colab import drive
import os

# Function to split genres and create new rows
def split_genres(row):
  if isinstance(row['Genre'], list):
    new_rows = []
    for i in range(len(row['Genre'])):
      new_row = row.copy()
      new_row['Genre'] = row['Genre'][i]
      new_rows.append(new_row)
    return new_rows
  else:
    return [row]

# Apply the function to split genres and create new rows
new_movie_df = pd.concat([pd.DataFrame(split_genres(row)) for _, row in movie_df.iterrows()], ignore_index=True)

# Filter out rows with genres not in the genres_list
new_movie_df = new_movie_df[new_movie_df['Genre'].isin(genres_list)]
````

````
new_movie_df
````

````
import matplotlib.pyplot as plt
import numpy as np

# Group the data by genre and count the occurrences
genre_counts = new_movie_df['Genre'].value_counts()

# Generate colors using a colormap (rainbow)
colors = plt.cm.rainbow(np.linspace(0, 1, len(genre_counts)))

# Set figure size for better readability
plt.figure(figsize=(10, 6))

# Create the line plot with rainbow colors
for i in range(len(genre_counts)):
    plt.plot(genre_counts.index[i:i+2], genre_counts.values[i:i+2], marker='o', color=colors[i], markersize=8, linewidth=2)

# Add labels and title with enhanced styling
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Most Genres of movies', fontsize=16, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=12)

# Add gridlines for better visual guidance
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()  # Adjust layout to fit everything
plt.show()
````

````
movie_df = movie_df.dropna(subset=['Duration','Rating'])
````

````
movie_df['Duration'] = movie_df['Duration'].str.replace('min', '').astype(int)
movie_df['Year'] = pd.to_numeric(movie_df['Year'], errors='coerce')
````

````
filtered_df = movie_df[(movie_df['Year'] >= 2000) & (movie_df['Year'] <= 2022)]
````

````
import seaborn as sns # Import the seaborn library and give it the alias 'sns'
import matplotlib.pyplot as plt
````

````
plt.figure(figsize=(10, 10))
scatter_plot = sns.scatterplot(x='Duration', y='Rating', data=filtered_df,
                               s=100, edgecolor='w', alpha=0.7, palette='coolwarm', hue='Year')
````

````
plt.figure(figsize=(10, 10))
scatter_plot = sns.scatterplot(x='Duration', y='Rating', data=filtered_df,
                               s=100, edgecolor='w', alpha=0.7, palette='coolwarm', hue='Year')

plt.title('Scatter Plot of Duration vs. Rating (2000-2022)', fontsize=24)
plt.xlabel('Duration (minutes)', fontsize=18)
plt.ylabel('Rating', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Calculate the mean duration and mean rating
mean_duration = filtered_df['Duration'].mean()
mean_rating = filtered_df['Rating'].mean()

# Now use the calculated values in the plt.text function
plt.text(mean_duration+5, plt.ylim()[1]-0.1, f'Mean Duration: {mean_duration:.2f} min',color='g', fontsize=12)
plt.text(plt.xlim()[1]-50, plt.ylim()[1]-0.1, f'Mean Rating: {mean_rating:.2f}',color='g', fontsize=12)
````
