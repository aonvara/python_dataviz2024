# import ข้อมูล indian director

```
#เอาข้อมูลใหม่เข้ามาที่เกี่ยวกับ director
data_director = pd.read_csv(os.path.join(path,'indian_director.csv'), encoding='latin-1')
data_director.head()
```

## ต่อแกน X


*   จับ 2 ตารางมาต่อกันเลย (merge)
*   เลือกมาเพิ่มเฉพาะบาง column (map)



```
#เอาเฉพาะบางข้อมูล
data_director_sm = data_director[['Name','Known For','Birth Date']]
data_director_sm
```

```
#เอาเฉพาะบางข้อมูลของ India
data_india_sm = data_india[['Name',	'Year',	'Duration',	'Genre',	'Rating',	'Votes',	'Director'	]]
data_india_sm
```

#merge 2 ตาราง

## เชื่อมแบบที่ 1

```
merged_df = pd.merge(data_director_sm, data_india_sm, left_on='Known For', right_on='Name')
merged_df
```

```
data_india[data_india['Name']=='Gangs of Wasseypur']
```

##ปัญหาการต่อตารางเป็นเพราะในอีกตารางมีค่า duplicates

```
data_director_sm['Known For'].map(data_director_sm.set_index('Name')['Duration'])
```

```
data_india_sm = data_india_sm.drop_duplicates(subset='Name',keep='last') #drop duplicate
```

```
data_india_sm[data_india_sm.duplicated(keep=False)] #check duplicate
```

```
data_india_sm[data_india_sm['Name']=='Zindagi Na Milegi Dobara']
```
